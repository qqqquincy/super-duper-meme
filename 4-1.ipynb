```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# 捷運車站每日各站進出量統計

```python
df = pd.read_csv('https://beta.data.taipei/api/getDatasetInfo/downloadResource?id=\
1d71c478-205f-42c5-8386-35f86d74fdd1&rid=0f7126ea-3598-468c-a1aa-bf43b2f483a8', encoding='BIG5') #大5編碼
```

```python
df.head(23)
```

```python
target = df[['日期','時段','科技大樓']].copy() #目標分析車站：科技大樓站
target.head()
```

```python
import locale
locale.setlocale(locale.LC_NUMERIC,'') 

target['科技大樓'] = target['科技大樓'].apply(locale.atof) #轉為float
target.head()
```

```python
time = np.arange(5,23)
Npassenger = np.zeros(time.shape)

for i in range(len(time)):
    t = target[target['時段']==time[i]]
    Npassenger[i] = round(t['科技大樓'].mean(), 0) #各時段的平均人數四捨五入到整數位
    
plt.rcParams['font.sans-serif'] = ['AppleGothic'] #中文字型

plt.figure(figsize=(10,5), dpi=80)
plt.bar(time, Npassenger, width=0.5, color='r')

plt.title('科技大樓站旅運量')
plt.xticks(time)

plt.xlabel('時段')
plt.ylabel('平均人數')

plt.show()
```
